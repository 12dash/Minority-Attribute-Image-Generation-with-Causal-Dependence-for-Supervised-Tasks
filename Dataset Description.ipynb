# Count plot for various datasets used in this project

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")
```

```python
df = pd.read_csv('dataset/celebA/attr.csv')
```

```python
label = 'Smiling'
attr = 'Male'
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data = df, x = label, hue =attr)
plt.ylabel('# Imgs')
plt.title('Natural Distribution')
plt.show()
```

```python
df_train = pd.read_csv('dataset/celebA/dear_train.csv')
df_test = pd.read_csv('dataset/celebA/dear_test.csv')
df_val = pd.read_csv('dataset/celebA/dear_val.csv')
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data = df_train, x = label, hue =attr)
plt.ylabel('# Imgs')
plt.title('Training Distribution Before Downsampling')
plt.show()

plt.figure(figsize=(5,3))
sns.countplot(data = df_test, x = label, hue =attr)
plt.ylabel('# Imgs')
plt.title('Testing Distribution')
plt.show()

plt.figure(figsize=(5,3))
sns.countplot(data = df_val, x = label, hue =attr)
plt.ylabel('# Imgs')
plt.title('Validation Distribution')
plt.show()
```

```python
df_down = df_val = pd.read_csv('dataset/celebA/dear_train_downsample_smile.csv')

plt.figure(figsize=(5,3))
sns.countplot(data = df_down, x = label, hue =attr)
plt.ylabel('# Imgs')
plt.title('Training Distribution after Downsampling')
plt.show()
```
