```python
folder = "dataset/celebA/"
img_folder = folder+'img_align_celeba/img_align_celeba/'

df = pd.read_csv(folder + "attr.csv")
df.head()
```

### Dear dataset split

```python
dear_train, dear_test = train_test_split(df, test_size = 0.2, random_state = 42)
dear_test, dear_val = train_test_split(dear_test, test_size = 0.5, random_state = 42)
print(f"# dear_train : {len(dear_train)} # dear_val : {len(dear_val)} # dear_test : {len(dear_test)}")
```

```python
folder_dir = 'dataset/celebA/'

dear_train.to_csv(f'{folder_dir}dear_train.csv', index=False)
dear_test.to_csv(f'{folder_dir}dear_test.csv', index=False)
dear_val.to_csv(f'{folder_dir}dear_val.csv', index=False)
```
